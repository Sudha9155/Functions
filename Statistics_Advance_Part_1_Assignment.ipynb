Q1. What is a random variable in probability theory?

Answer:
A random variable is a numerical value assigned to the outcomes of a random experiment. It represents uncertain outcomes quantitatively. For example, the number obtained when rolling a die is a random variable.

Q2. What are the types of random variables?

Answer:
There are two main types of random variables:
1. Discrete Random Variable – Takes countable values (e.g., number of students in a class).
2. Continuous Random Variable – Takes any value within a range (e.g., height or weight of a person).

Q3. What is the difference between discrete and continuous distributions?

Answer:
A discrete distribution describes probabilities of countable outcomes, while a continuous distribution describes probabilities over a continuous range of values. Discrete variables are counted, whereas continuous variables are measured.

Q4. What are probability distribution functions (PDF)?

Answer:
A Probability Distribution Function describes how probabilities are distributed over the values of a random variable. For continuous variables, the PDF indicates the relative likelihood of different outcomes.

Q5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?

Answer:
A PDF gives the probability density at a particular value, whereas a CDF gives the cumulative probability that a random variable is less than or equal to a specific value.

Q6. What is a discrete uniform distribution?

Answer:
A discrete uniform distribution is a probability distribution where all possible outcomes have equal probability. For example, each face of a fair six-sided die has a probability of 1/6.

Q7. What are the key properties of a Bernoulli distribution?

Answer:
A Bernoulli distribution has only two possible outcomes: success (1) and failure (0). It is characterized by a single probability parameter p, where P(success)=p and P(failure)=1−p.

Q8. What is the binomial distribution, and how is it used in probability?

Answer:
The binomial distribution models the number of successes in a fixed number of independent Bernoulli trials. It is commonly used in quality control, surveys, and medical studies.

Q9. What is the Poisson distribution and where is it applied?

Answer:
The Poisson distribution models the number of events occurring within a fixed interval of time or space. It is used for call arrivals, website visits, accidents, and system failures.

Q10. What is a continuous uniform distribution?

Answer:
A continuous uniform distribution is a probability distribution where all values within a specified interval are equally likely to occur.

Q11. What are the characteristics of a normal distribution?

Answer:
A normal distribution is bell-shaped, symmetric about the mean, and characterized by its mean and standard deviation. The mean, median, and mode are equal in a normal distribution.

Q12. What is the standard normal distribution, and why is it important?

Answer:
The standard normal distribution is a normal distribution with mean 0 and standard deviation 1. It is important because it allows comparison of different datasets using Z-scores.

Q13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?

Answer:
The Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as sample size increases, regardless of the population distribution. It forms the basis of statistical inference.

Q14. How does the Central Limit Theorem relate to the normal distribution?

Answer:
The CLT explains why many sampling distributions become approximately normal when the sample size is sufficiently large, even if the original data is not normally distributed.

Q15. What is the application of Z statistics in hypothesis testing?

Answer:
Z-statistics are used to determine whether a sample mean significantly differs from a population mean when the population variance is known or the sample size is large.

Q16. How do you calculate a Z-score, and what does it represent?

Answer:
A Z-score is calculated as:

Z = (X − μ) / σ

where X is the observation, μ is the mean, and σ is the standard deviation. It represents how many standard deviations a value is from the mean.

Q17. What are point estimates and interval estimates in statistics?

Answer:
A point estimate provides a single value estimate of a population parameter, whereas an interval estimate provides a range of values within which the parameter is likely to lie.

Q18. What is the significance of confidence intervals in statistical analysis?

Answer:
Confidence intervals provide a range of plausible values for a population parameter and indicate the reliability and precision of an estimate.

Q19. What is the relationship between a Z-score and a confidence interval?

Answer:
Z-scores are used to calculate confidence intervals. The confidence interval is determined by adding and subtracting the margin of error based on the Z-value.

Q20. How are Z-scores used to compare different distributions?

Answer:
Z-scores standardize data, allowing values from different distributions to be compared on the same scale.

Q21. What are the assumptions for applying the Central Limit Theorem?

Answer:
The observations should be independent, the sample should be randomly selected, and the sample size should be sufficiently large.

Q22. What is the concept of expected value in a probability distribution?

Answer:
Expected value is the long-run average outcome of a random variable. It is calculated as the weighted average of all possible outcomes using their probabilities.

Q23. How does a probability distribution relate to the expected outcome of a random variable?

Answer:
A probability distribution assigns probabilities to all possible values of a random variable. The expected outcome is calculated from this distribution and represents the average value expected over many repetitions.

```
# ==========================================
# Q6. Write a Python program to calculate and plot the cumulative
# distribution function (CDF) of a discrete uniform distribution.
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 7)
pmf = np.ones(len(x)) / len(x)
cdf = np.cumsum(pmf)

plt.figure(figsize=(6,4))
plt.step(x, cdf, where='mid')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Values")
plt.ylabel("Cumulative Probability")
plt.grid(True)
plt.show()


# ==========================================
# Q7. Generate a continuous uniform distribution using NumPy
# and visualize it.
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

data = np.random.uniform(0, 10, 1000)

plt.figure(figsize=(6,4))
plt.hist(data, bins=30, edgecolor='black')
plt.title("Continuous Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


# ==========================================
# Q8. Simulate data from a normal distribution and plot its histogram.
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

normal_data = np.random.normal(loc=0, scale=1, size=1000)

plt.figure(figsize=(6,4))
plt.hist(normal_data, bins=30, edgecolor='black')
plt.title("Normal Distribution Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


# ==========================================
# Q9. Write a Python function to calculate Z-scores from a dataset
# and plot them.
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

data = np.array([10, 12, 15, 18, 20, 22, 25])

def z_scores(dataset):
    mean = np.mean(dataset)
    std = np.std(dataset)
    return (dataset - mean) / std

z = z_scores(data)

print("Dataset:", data)
print("Z-Scores:", z)

plt.figure(figsize=(6,4))
plt.plot(data, z, marker='o')
plt.title("Z-Score Plot")
plt.xlabel("Data Values")
plt.ylabel("Z-Score")
plt.grid(True)
plt.show()


# ==========================================
# Q10. Implement the Central Limit Theorem (CLT)
# using Python for a non-normal distribution.
# ==========================================

import numpy as np
import matplotlib.pyplot as plt

sample_means = []

for i in range(1000):
    sample = np.random.exponential(scale=2, size=30)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(6,4))
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("Central Limit Theorem Demonstration")
plt.xlabel("Sample Mean")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```
# =====================================================
# Q15. Simulate multiple samples from a normal distribution
# and verify the Central Limit Theorem.
# =====================================================

import numpy as np
import matplotlib.pyplot as plt

sample_means = []

for i in range(1000):
    sample = np.random.normal(loc=50, scale=10, size=30)
    sample_means.append(np.mean(sample))

plt.figure(figsize=(6,4))
plt.hist(sample_means, bins=30, edgecolor='black')
plt.title("CLT Verification using Normal Samples")
plt.xlabel("Sample Means")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()


# =====================================================
# Q16. Write a Python function to calculate and plot
# the standard normal distribution (mean=0, std=1).
# =====================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(6,4))
plt.plot(x, y)
plt.title("Standard Normal Distribution")
plt.xlabel("X")
plt.ylabel("PDF")
plt.grid(True)
plt.show()


# =====================================================
# Q17. Generate random variables and calculate their
# corresponding probabilities using the binomial distribution.
# =====================================================

import numpy as np
from scipy.stats import binom

n = 10
p = 0.5

x = np.arange(0, n + 1)
probabilities = binom.pmf(x, n, p)

print("Random Variables:", x)
print("Probabilities:", probabilities)


# =====================================================
# Q18. Write a Python program to calculate the Z-score
# for a given data point and compare it to a standard
# normal distribution.
# =====================================================

import numpy as np
from scipy.stats import norm

data_point = 85
mean = 70
std = 10

z_score = (data_point - mean) / std

print("Z-Score:", z_score)
print("Probability:", norm.cdf(z_score))


# =====================================================
# Q19. Implement hypothesis testing using Z-statistics
# for a sample dataset.
# =====================================================

import numpy as np
from scipy.stats import norm

sample_mean = 105
population_mean = 100
sigma = 15
n = 50

z = (sample_mean - population_mean) / (sigma / np.sqrt(n))
p_value = 2 * (1 - norm.cdf(abs(z)))

print("Z Statistic:", z)
print("P-Value:", p_value)

if p_value < 0.05:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")


# =====================================================
# Q20. Create a confidence interval for a dataset
# using Python and interpret the result.
# =====================================================

import numpy as np
from scipy import stats

data = np.array([10,12,14,15,18,20,22,25,28,30])

confidence = 0.95

mean = np.mean(data)
sem = stats.sem(data)

interval = stats.t.interval(
    confidence,
    len(data)-1,
    loc=mean,
    scale=sem
)

print("Mean:", mean)
print("95% Confidence Interval:", interval)


# =====================================================
# Q21. Generate data from a normal distribution, then
# calculate and interpret the confidence interval for its mean.
# =====================================================

import numpy as np
from scipy import stats

data = np.random.normal(loc=50, scale=10, size=100)

mean = np.mean(data)
sem = stats.sem(data)

ci = stats.t.interval(
    0.95,
    len(data)-1,
    loc=mean,
    scale=sem
)

print("Sample Mean:", mean)
print("95% Confidence Interval:", ci)
```
