Q1. What is the difference between a function and a method in Python?

Answer:

Example:

Q3. What are the different ways to define and call a function in Python?

Answer:

Q5. What are iterators in Python and how do they differ from iterables?

Answer:
An iterable is an object that can be looped over, such as a list or tuple. An iterator is an object that returns one item at a time using the next() function.

Example:
nums = [1, 2, 3]
it = iter(nums)

print(next(it))

⸻

Q6. Explain the concept of generators in Python and how they are defined.

Answer:
Generators are special functions that use the yield keyword to produce values one at a time. They save memory because values are generated only when needed.

Example:
def count():
yield 1
yield 2
yield 3

for i in count():
print(i)

⸻

Q7. What are the advantages of using generators over regular functions?

Answer:
Generators use less memory, generate values on demand, and are useful when working with large datasets.

Example:
def numbers():
for i in range(5):
yield i

for num in numbers():
print(num)

⸻

Q8. What is a lambda function in Python and when is it typically used?

Answer:
A lambda function is a small anonymous function written in a single line. It is commonly used for short operations with functions like map(), filter(), and sorted().

Example:
square = lambda x: x*x
print(square(4))

Q9. Explain the purpose and usage of the map() function in Python.

Answer:
The map() function applies a specified function to each item of an iterable and returns a map object containing the results.

Example:
nums = [1, 2, 3, 4]

result = list(map(lambda x: x*x, nums))

print(result)

Q10. What is the difference between map(), reduce(), and filter() functions in Python?

Answer:
map() applies a function to every element of an iterable.
filter() selects elements that satisfy a condition.
reduce() combines all elements into a single value.

Example:
from functools import reduce

nums = [1, 2, 3, 4]

print(list(map(lambda x: x*2, nums)))
print(list(filter(lambda x: x%2==0, nums)))
print(reduce(lambda x, y: x+y, nums))

Q1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```
def sum_even(numbers):
    return sum(num for num in numbers if num % 2 == 0)

nums = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_even(nums))
```

Q4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False

    return True

num = 97

if is_prime(num):
    print(num, "is Prime")
else:
    print(num, "is Not Prime")
```

Q5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```
class Fibonacci:
    def __init__(self, n):
        self.n = n
        self.a = 0
        self.b = 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n:
            raise StopIteration

        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

for num in Fibonacci(10):
    print(num)
```

Q6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

for value in powers_of_two(8):
    print(value)
```

Q7. Implement a generator function that reads a file line by line and yields each line as a string.

```
def read_file(filename):
    with open(filename, "r") as file:
        for line in file:
            yield line.strip()

for line in read_file("sample.txt"):
    print(line)
```

Q8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```
data = [
    ("A", 5),
    ("B", 2),
    ("C", 8),
    ("D", 1)
]

sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

Q9. Write a Python program that uses map() to convert a list of temperatures from Celsius to Fahrenheit.

```
celsius = [0, 10, 20, 30, 40]

fahrenheit = list(
    map(lambda c: (c * 9/5) + 32, celsius)
)

print(fahrenheit)
```

Q10. Create a Python program that uses filter() to remove all the vowels from a given string.

```
text = "Sudhanshu"

result = "".join(
    filter(
        lambda ch: ch.lower() not in "aeiou",
        text
    )
)

print(result)
```

Q11. Imagine an accounting routine used in a book shop. It works on a list with sublists. Write a Python program using lambda and map that returns a list with 2-tuples. Each tuple consists of the order number and the product of the price per item and quantity. Increase the product by €10 if the order value is less than €100.

```
orders = [
    (34587, "Learning Python, Mark Lutz", 4, 40.95),
    (98762, "Programming Python, Mark Lutz", 5, 56.80),
    (77226, "Head First Python, Paul Barry", 3, 32.95),
    (88112, "Einführung in Python3, Bernd Klein", 3, 24.99)
]

result = list(
    map(
        lambda item: (
            item[0],
            item[2] * item[3] + 10
            if item[2] * item[3] < 100
            else item[2] * item[3]
        ),
        orders
    )
)

print(result)
```
